트리!! 
- Q) 왜 이렇게 그린 그래프는 트리일까? -> 사이클이 없다는 것을 증명해야함
- 사이클이 있다면 (선을 긋는다) -> 가는 경로가 2개! 가중치가 다름..! 그럼 적은 것을 그려야함! 따라서 같아야 그릴 수 있음.. 그러나 같으면 나머지 선을 안그으면 됨

- 증명 : 최단거리는 항상 1개밖에 없으니까
- 여러개 있다면 그냥 안그으면 됨

- 다익스트라 알고리즘은 최단경로 트리 (Shortest-path Tree) 를 구하는 것!


```python
import sys
sys.setrecursionlimit(100000)

def getShortest(graph, start, end) :
    '''
    graph가 주어질 때, start부터 end까지의 최단거리를 반환하는 함수를 작성하세요.
    
    graph[0] = [(1, 4), (2, 3), (3, 7), (4, 3)]
    0이랑 인접한 것
    1번 노드까지 가는데 길이 4
    
    1. 노드까지의 거리를 저장하는 배열 초기화 (Table)
        Table[i] : i까지 오는데 걸리는 최단거리
    2. 해당 노드를 방문했는지에 대한 여부 (visited)
        visited[i] : i를 방문했으면 True, 아니면 False
    3. Table에서 가장 작은 값 찾기
        (단, 지금까지 방문하지 않은 정점)
    4. 뻗기
    5. 다시 3으로 돌아가기
    
    Q. 몇번 반복해야 할까요? 3 - 5 를? n번!(정점 개수만큼)
    
    '''
    n = len(graph)
    infinity = 989812312312
    
    Table = [infinity for i in range(n)]
    visited = [False for i in range(n)]

    Table[start] = 0
    
    for i in range(n):
        '''
        1. 최솟값 찾기 (단, 방문안한 애들 중에서)
        2. 뻗기
        '''
        
        myMin = infinity # 정점의 거리
        myMinIdx = -1 # 정점의 번호
        
        for j in range(n):
            if Table[j] < myMin and visited[j] == False:
                myMin = Table[j]
                myMinIdx = j
                
        cur = myMinIdx
        
        visited[cur] = True
        
        for j in range(len(graph[cur])):
            myNext = graph[cur][j][0]
            myCost = graph[cur][j][1]
            
            if Table[cur] + myCost < Table[myNext]:
                Table[myNext] = Table[cur] + myCost
                
    return Table[end]

def main():
    '''
    Do not change this code
    '''

    line = [int(x) for x in input().split()]

    n = line[0]
    m = line[1]
    start = line[2]
    end = line[3]

    graph = [ [] for i in range(n) ]

    for i in range(m) :
        line = [int(x) for x in input().split()]

        graph[line[0]].append((line[1], line[2]))
        graph[line[1]].append((line[0], line[2]))

    print(getShortest(graph, start, end))

if __name__ == "__main__":
    main()
```

## 스패닝 트리

- 그래프의 모든 노드를 포함하는 트리

### 최소 스패닝 트리 ( Minumum Spanning Tree ) - 최소 신장 트리
- 그래프의 모든 노드를 포함하는 트리 중 간선의 가중치의 합이 최소인 트리
- 짧은 것부터 그으면됨!

#### Cut property
- 두 개의 MST를 연결하여 하나의 MST를 만들어야 한다!
- 가장 가중치가 작은 것을 이어야 합니다

### 크루스칼 알고리즘 ( Kruskal's Algorithm )
- 각 노드를 하나의 MST로 보고 시작
- 하나의 트리를 하나의 그룹으로 파악할 수 있는가?


### Disjoint Set
- 두 정점을 하나의 그룹으로 묶어라
- 두 정점이 같은 그룹에 속하는가?

하나의 트리 
    1. 트리의 루트는 트리를 대표하는 정점
    2. 트리 안에 있는 정점들은 같은 그룹이다
    
    
- 하나의 그룹인지 확인하려면, 각각의 루트를 확인해보면 됨!
- 합치는 경우는 루트를 변경하면 됨 (상속 같은 느낌)

```python
import sys
sys.setrecursionlimit(100000)

def find(parent, x):
    '''
    x의 루트를 반환하는 함수
    '''
    if parent[x] == x:
        return x
    else:
        return find(parent, parent[x])

def union(parent, a, b):
    '''
    a와 b가 서로 다른 그룹이라면 True 및 합치기
    이미 같은 그룹이라면 False
    
    1. a의 루트 구하기
    2. b의 루트 구하기
    3. 두 루트가 같니? 같다면 False
    4. 다르다면, 한 루트의 부모를 다른 루트로 정하기
        True 반환
    '''

    a = find(parent, a)
    b = find(parent, b)
    
    # a와 b는 루트
    
    if a == b:
        return False
    else:
        parent[b] = a
        return True
        
def getMST(graph) :
    '''
    graph가 주어질 때, 그 최소 비용 신장트리의 간선 가중치의 합을 반환하는 함수를 작성하세요.
    
    graph[0] = [(2,3), (1, 4), (7, 8)]
    
    1. 간선을 가중치가 적은 순서대로 정렬
    2. 그 적은 순서대로 하나씩 연결
    3. 연결할 떄 같은 트리에 없을때만 연결
    
    '''
    
    edges = [] # 간선을 저장 (정점 1, 정점 7, 가중치 3)
    
    n = len(graph)
    
    for i in range(n):
        for j in range(len(graph[i])):
            edges.append((i, graph[i][j][0], graph[i][j][1]))
    
    # edges를 가중치 순서대로 정렬
    
    #edges.sort() 정렬 안됨
    
    def compare(x):
        return x[2]
        
    edges.sort(key=compare)
    
    result = 0
    
    parent = [ i for i in range(n) ]
    
    for e in edges:
        if union(parent, e[0],e[1]) == True:
            result += e[2]
    
    
    return result

def main():
    '''
    Do not change this code
    '''

    line = [int(x) for x in input().split()]

    n = line[0]
    m = line[1]

    graph = [ [] for i in range(n) ]

    for i in range(m) :
        line = [int(x) for x in input().split()]

        graph[line[0]].append((line[1], line[2]))
        graph[line[1]].append((line[0], line[2]))

    print(getMST(graph))

if __name__ == "__main__":
    main()
```
